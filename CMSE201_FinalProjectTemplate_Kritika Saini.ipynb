```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from pandas.plotting import lag_plot
import numpy as np
```

```python
beer = pd.read_csv("beer1.csv") 
beer.head() 
```

Here is a list of all the years i collected data of and we are going to study the prices of these many years.

```python
Year = beer["Year"]
Region = beer['Region']
```

```python
YearList = beer["Year"]
print(YearList)
```

Here is a list of average price of beer in a particular year.

```python
PriceList = [beer["AVG_Price"]]
print(PriceList)
```

Here I have made a histogram of all the average prices

```python
Prices = beer["AVG_Price"]
plt.rcParams["figure.figsize"] = [30,30]
plt.hist(Prices, color='pink')
plt.title("Average Prices")
plt.grid()
```

```python
plt.plot(Prices, color="blue")
plt.rcParams["figure.figsize"] = [30,30]
plt.title("Average Prices")
plt.grid()
```

Here I have calculated the mean, median, minimum and maximum value of the average price list and so we can calculate the differences after using numpy to help us calculate these. 

```python
mean = np.mean(Prices)
print("Mean of Average Prices of Beer: ", mean)
#since no price was repeated and every year had a unique price, the median will come out to be Not a Number(nan)
Median = np.median(Prices)
print("Median of Average Prices of Beer: ", Median)

Minimum = np.min(Prices)
print("Minimum of Average Prices of Beer: ", Minimum)

Maximum = np.max(Prices)
print("Maximum of Average Prices of Beer: ", Maximum)
print()
print("The Range is: ", Maximum-Minimum)
print()
print("Difference from Maximum to Minimum: ", Maximum-Minimum)
print("Difference from Maximum to Mean: ", Maximum-mean)
print()
```

So here I have used seaborn to make a grid graph and understand the prices better.

We can clearly see from the graph above that the number of breweries straight up increased over the years.
